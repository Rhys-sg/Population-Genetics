```python
# 2 loci with 2 alleles each (Not HW equilibrium)

generations=200
growth_rate=1
drift=0
carrying_capacity = 100000
bottleneck_yr=None
bottleneck_pop=None
covariance=0
small_pop_inbreeding=0

genotype_data = {
    (('A1', 'A2'), ('B1', 'B1')): {'male': 100, 'female' : 100, 'fitness' : 1.0},
    (('A1', 'A2'), ('B1', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 1.0},
    (('A1', 'A2'), ('B2', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 1.0},
    (('A2', 'A2'), ('B1', 'B1')): {'male': 100, 'female' : 100, 'fitness' : 0.8},
    (('A2', 'A2'), ('B1', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 0.8},
    (('A2', 'A2'), ('B2', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 0.8},
    (('A1', 'A1'), ('B1', 'B1')): {'male': 100, 'female' : 100, 'fitness' : 0.7},
    (('A1', 'A1'), ('B1', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 0.7},
    (('A1', 'A1'), ('B2', 'B2')): {'male': 100, 'female' : 100, 'fitness' : 0.7},
}

mutations = [{('A1', 'A2'): 0.0, ('A2', 'A1'): 0.0,},
             {('B1', 'B2'): 0.0, ('B2', 'B1'): 0.0}]

pop_gen(
    generations,
    genotype_data,
    growth_rate,
    carrying_capacity,
    drift,
    mutations,
    bottleneck_yr,
    bottleneck_pop,
    covariance,
    small_pop_inbreeding
)
```
