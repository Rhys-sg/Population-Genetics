```python
# 2 locus, 3 alleles each (Randomly generated, not HW equilibrium)

generations=200
growth_rate=0.1
drift=0
carrying_capacity = 100000
bottleneck_yr=100
bottleneck_pop=50000

# Randomly generate genotype data
num_loci = 2
num_alleles = 3
population_size = 10000
random_seed = None
genotype_counts, genotype_fitness = generate_genotype_data(num_loci, num_alleles, population_size, random_seed)

mutations=None

next_genotype_frequencies(generations,
                          genotype_counts,
                          genotype_fitness,
                          growth_rate,
                          carrying_capacity,
                          drift,
                          mutations,
                          bottleneck_yr,
                          bottleneck_pop)

# Print genotype counts and fitness
print("Genotype Counts:")
for genotype, count in genotype_counts.items():
    print(f"{genotype}: {count}")

print("\nGenotype Fitness:")
for genotype, fitness in genotype_fitness.items():
    print(f"{genotype}: {fitness}")
```
